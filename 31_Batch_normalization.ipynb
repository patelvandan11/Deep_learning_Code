```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense ,BatchNormalization
from tensorflow.keras.models import Sequential
```

```python
import pandas as pd
df=pd.read_csv('concentric_circles.csv')
```

```python
model = Sequential()
model.add(Dense(3, input_dim=2, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(2, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
df.head()
```

```python
plt.scatter(df['Feature_1'],df['Feature_2'],c=df['Label'])
```

```python
X = df.iloc[:,0:2].values
y = df.iloc[:,-1].values
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
h1=model.fit(X,y,epochs=200,validation_split=0.2)
```

```python
model = Sequential()

model.add(Dense(2,activation='relu',input_dim=2))
model.add(Dense(2,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
h2 = model.fit(X,y,epochs=200,validation_split=0.2)
```

```python
plt.plot(h1.history['val_accuracy'], color='black', label='Model 1') # with Normalization
plt.plot(h2.history['val_accuracy'], color='green', label='Model 2') # without Normalization
plt.xlabel('Epochs')
plt.ylabel('Validation Accuracy')
plt.legend()
plt.plot(h2.history['val_accuracy'],color='green')
```
